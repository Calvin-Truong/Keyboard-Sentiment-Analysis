```python
import nltk
from nltk import WordNetLemmatizer
import numpy as np
```

```python
cps=np.array(df['Comment'])
```

```python
stop_words=nltk.corpus.stopwords.words('english')
wpt=nltk.WordPunctTokenizer()
```

```python
 for x in stop_words:
        if x==['no','nor','not']:
            stop_words.remove(x)
        if bool(re.search('n\'t',x)):
            stop_words.remove(x)
```

```python
stop_words
```

```python
def normalize_docuement (doc):
    #special char  and number and handling ' for negative words
    doc=re.sub('\'', '',doc) 
    doc=re.sub('[^a-zA-Z]', ' ',doc)
    
    #single char
    doc=re.sub(r'\s+[a-zA-Z]\s+',' ',doc) 
    doc=re.sub(r'^[a-zA-Z]\s+', ' ', doc, re.I|re.A) 
    
    doc=re.sub(r'\s+',' ',doc).lower() # extra white space and lowercase
    #tokenize document
    tokens=wpt.tokenize(doc)
    #filter stopwords out of document
    filtered_tokens=[token for token in tokens if token not in stop_words]
   # doc=' '.join(filtered_tokens)
    
    #lemmentize
    lemmatizer = WordNetLemmatizer()
    newtokens = [lemmatizer.lemmatize(filtered_token) for filtered_token in filtered_tokens]
    #re-create document
    doc=' '.join(newtokens)
    
    return doc
```

```python
normalize_corpus=np.vectorize(normalize_docuement) 

normalized_cps=normalize_corpus(cps)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

LDA = LatentDirichletAllocation(n_components=5,random_state=42)
```

```python
# tfidf

from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer (min_df=10, max_df=0.95, norm='l2', 
                  use_idf=True, smooth_idf=True)
tv_matrix=tv.fit_transform(normalized_cps)
tv_matrix=tv_matrix.toarray()
vocab=tv.get_feature_names()
pd.DataFrame (np.round(tv_matrix,2), columns=vocab)
```

```python
LDA.fit(tv_matrix)
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 10 WORDS FOR TOPIC #{index}')
    print([tv.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
#BOW
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(min_df=10, max_df=0.95)

cv_matrix=cv.fit_transform(normalized_cps)

cv_matrix=cv_matrix.toarray()

#get all unique words in the corpus
vocab=cv.get_feature_names()
#show document feature matrix
pd.DataFrame(cv_matrix,columns=vocab)
```

```python
LDA.fit(cv_matrix)
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 10 WORDS FOR TOPIC #{index}')
    print([cv.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
LDA.fit(tv_matrix)
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 10 WORDS FOR TOPIC #{index}')
    print([tv.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
bv = CountVectorizer(ngram_range=(3,3),min_df=2,max_df=0.95)
bv_matrix=bv.fit_transform(normalized_cps)
bv_matrix=bv_matrix.toarray()
vocab=bv.get_feature_names()
pd.DataFrame(bv_matrix,columns=vocab)
```

```python
#bv -3grams
LDA.fit(bv_matrix)
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 10 WORDS FOR TOPIC #{index}')
    print([bv.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
bv = CountVectorizer(ngram_range=(4,4),min_df=2,max_df=0.95)
bv_matrix=bv.fit_transform(normalized_cps)
bv_matrix=bv_matrix.toarray()
vocab=bv.get_feature_names()
pd.DataFrame(bv_matrix,columns=vocab)
```

```python
#bv -4grams
LDA.fit(bv_matrix)
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 10 WORDS FOR TOPIC #{index}')
    print([bv.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
topic_results= LDA.transform(bv_matrix)
```

```python
df['Topic'] = topic_results.argmax(axis=1)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df['scores'] = df['Comment'].apply(lambda review: sid.polarity_scores(review))
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])
df['comp_category'] = df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')

df.head()
```
